```python
import pandas as pd
```

```python
economy_data = pd.read_csv("https://storage.googleapis.com/covid19-open-data/v3/economy.csv")
```

```python
print(economy_data.info)
economy_data.head()
```

```python
index_data = pd.read_csv("country_credentials.csv")
```

```python
economy_data = economy_data[economy_data["location_key"].isin(index_data["country_code"])]
economy_data.reset_index(inplace=True, drop=True)
display(economy_data)
```

```python
economy_data.fillna(0, inplace=True)
numeric_cols = [
"gdp_usd",	
"gdp_per_capita_usd",	
"human_capital_index",
]
for _,col in enumerate(numeric_cols):
    economy_data[economy_data[col] < 0] = 0
display(economy_data)
```
