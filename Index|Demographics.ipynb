```python
import pandas as pd
demographics_data = pd.read_csv("https://storage.googleapis.com/covid19-open-data/v3/demographics.csv")
```

```python
demographics_data.fillna(0, inplace=True)
print(demographics_data.info())

demographics_data.head()
```

```python
# next thing to do here is to remove repetitions are multiple entries, because as observed the AF has total population of afghanistan and AF_something has some other information which isnt very required in this case, hence we'll have to look for somethings here and in the end this column integrated with the previous we will get a new csv


numeric_cols = [
    "population",
    "population_male",
	"population_female",
	"population_rural",
	"population_urban",
	"population_largest_city",
	"population_clustered",
	"population_density",
    "human_development_index",
    "population_age_00_09",
	"population_age_10_19",
	"population_age_20_29",
	"population_age_30_39",
	"population_age_40_49",
	"population_age_50_59",
	"population_age_60_69",
	"population_age_70_79",
	"population_age_80_and_older",
]
# demographics_data[numeric_cols] = demographics_data[numeric_cols].apply(pd.to_numeric, errors='coerce')
# Replace negative values with 0
for _,col in enumerate(numeric_cols):
    demographics_data[demographics_data[col] < 0] = 0
display(demographics_data)
demographics_data = demographics_data[demographics_data["location_key"].isin(index_data["country_code"])]
demographics_data = demographics_data.rename(columns={'location_key': 'country_code'})
```

```python
demographics_data.reset_index(drop=True, inplace=True)
display(demographics_data)
```

```python
demographics_data = pd.merge(index_data, demographics_data, on=["country_code"])
```

```python
display(demographics_data)
```

```python
demographics_data.to_csv("demographics_data.csv",encoding="utf-8")
```
